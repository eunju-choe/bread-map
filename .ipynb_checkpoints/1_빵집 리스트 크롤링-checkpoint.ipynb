```python
import time
import random
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import warnings

warnings.filterwarnings('ignore')
```

```python
def get_text(result, select):
    return result.select(select)[0].text
```

```python
df_add = pd.read_csv('address.csv')
df_add.head()
```

address.csv 파일에는 시 / 구 의 형태로 부산의 16개 구/군 데이터가 저장되어있음. 이를 불러와서 크롤링에 활용하고자함

크롤링 단계별로 살펴본 후, 마지막에 반복문으로 코드를 작성해보겠다.

```python
# 드라이버 실행
driver_path = 'chromedriver.exe'  # 드라이버 위치 지정
driver = webdriver.Chrome(driver_path)

# 카카오맵으로 이동
url = 'https://map.kakao.com/'
driver.get(url)
```

크롬 드라이버를 실행해서 카카오맵으로 이동한다.

```python
# 검색어 입력
keyword = '부산 남구 빵집'
element = driver.find_element(By.ID, 'search.keyword.query')
element.send_keys(keyword)

# 검색 버튼 클릭
submit_button = driver.find_element(By.ID,'search.keyword.submit')
driver.execute_script("arguments[0].click();", submit_button)
```

위의 코드를 이용하여 검색창에 키워드를 입력한다.
지금은 키워드를 직접 설정하였지만 반복문으로 할 때에는 df_add에서 순서대로 불러오는 방식을 활용할 것이다.

카카오맵에서는 인기도순 정렬이 가능하다. 인기도순으로 정렬하여 빵집 리스트를 추출하고자 한다. 인기도순을 클릭하기 위해 코드를 작성하면 오류가 발생한다.

이 지도 설정 표시 때문이다. 그래서 이걸 없애주기 위해 지도 화면을 한 번 클릭하는 코드를 먼저 실행해야한다.

```python
# 지도 설정 지우기
actions = ActionChains(driver)
actions.move_by_offset(500, 500).click().perform()

# 인기도순 정렬
element = driver.find_element(By.XPATH,'/html/body/div[5]/div[2]/div[1]/div[7]/div[5]/div[1]/ol/li[2]/a')
element.click()
```

지도 설정 창이 사라지고 인기도순으로 정렬되었다. 이제 검색 결과를 저장하면 된다. (우리 데이터 수집 규칙 설명)

```python
get_list = ['[data-id="screenOutName"]', '[data-id="scoreNum"]', '[data-id="numberofscore"]',
           '[data-id="numberofreview"]', '[data-id="address"]']

# 7페이지까지 데이터 수집
rank, num_page = 1, 1
search_result = []

for _ in range(8):
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    soup = soup.select('.placelist > li')

    # 페이지 내 데이터 수집
    for i in range(len(soup)):        
        result = [rank]
        for get in get_list:
            result.append(get_text(soup[i], get))
        rank += 1
        search_result.append(result)

    # page 이동
    if num_page == 5:
        page_button = driver.find_element(By.ID, 'info.search.page.next')
        num_page = 1
    else : 
        num_page += 1
        page_button = driver.find_element(By.ID, f'info.search.page.no{num_page}')

    driver.execute_script("arguments[0].click();", page_button)
    time.sleep(random.uniform(1, 5))
```

```python
columns = ['순위', '상호', '별점', '별점 리뷰 개수', '리뷰 개수', '주소']
df = pd.DataFrame(search_result, columns = columns)
```

이 df를 구/군별로 저장할 것이다. 이 코드를 반복하기 위해 정리하면 다음과 같다.

```python
is_first_iteration = True
get_list = ['[data-id="screenOutName"]', '[data-id="scoreNum"]', '[data-id="numberofscore"]',
           '[data-id="numberofreview"]', '[data-id="address"]']
columns = ['순위', '상호', '별점', '별점 리뷰 개수', '리뷰 개수', '주소']



# 드라이버 실행
driver_path = 'chromedriver.exe' # 드라이버 위치
driver = webdriver.Chrome(driver_path)

# 카카오맵으로 이동
url = 'https://map.kakao.com/'
driver.get(url)
time.sleep(random.uniform(1, 5))

for add1, add2 in zip(df_add['add1'], df_add['add2']):
    search_result = []
    
    # 검색어 입력
    keyword = (' ').join([add1[:2], add2, "빵집"])
    element = driver.find_element(By.ID, 'search.keyword.query')
    element.clear()
    element.send_keys(keyword)
    time.sleep(random.uniform(1, 5))
    
    # 검색 버튼 클릭
    submit_button = driver.find_element(By.ID,'search.keyword.submit')
    driver.execute_script("arguments[0].click();", submit_button)
    time.sleep(random.uniform(1, 5))
    
    # 지도 설정 창 지우기
    if is_first_iteration:
        actions = ActionChains(driver)
        actions.move_by_offset(500, 500).click().perform()
        time.sleep(random.uniform(1, 5))
        is_first_iteration = False
    
    # 인기도순 정렬
    element = driver.find_element(By.XPATH,'/html/body/div[5]/div[2]/div[1]/div[7]/div[5]/div[1]/ol/li[2]/a')
    element.click()
    time.sleep(random.uniform(1, 5))
    
    # 7페이지까지 데이터 수집
    rank = 1
    num_page = 1
    
    for _ in range(8):
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        soup = soup.select('.placelist > li')
        
        # 페이지 내 데이터 수집
        for i in range(len(soup)):        
            result = [rank]
            for get in get_list:
                result.append(get_text(soup[i], get))
            rank += 1
            search_result.append(result)

        # page 이동
        if num_page == 5:
            page_button = driver.find_element(By.ID, 'info.search.page.next')
            num_page = 1
        else : 
            num_page += 1
            page_button = driver.find_element(By.ID, f'info.search.page.no{num_page}')
        
        driver.execute_script("arguments[0].click();", page_button)
        time.sleep(random.uniform(1, 5))
        
    df = pd.DataFrame(search_result, columns = columns)
    dirt = (' ').join([add1, add2, '조회 결과'])
    df.to_csv('result/'+ dirt + '.csv', index=False)
    
driver.quit()
```
